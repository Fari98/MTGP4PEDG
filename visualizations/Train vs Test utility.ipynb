```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../log/compare_nlt_20250527.csv', names = ['dataset', 'train', 'validation', 'seed', 'position_in_front'])
```

```python
data['position_in_front'] = data['position_in_front'].apply(lambda x : int(x.split('.')[0]))
```

```python
datasets = data.dataset.unique()
```

```python
datasets
```

```python
for dataset in datasets:
    grouped = data[data['dataset'] == dataset][['dataset', 'train', 'validation']].groupby('dataset').median()
    plt.boxplot([grouped.train.values, grouped.validation.values], labels = ['train', 'validation'])

    plt.title(dataset)
    plt.show()
```
